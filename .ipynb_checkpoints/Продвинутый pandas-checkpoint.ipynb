# Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

1. если источник traffic_source равен yandex или google, то в source_type ставится organic
2. для источников paid и email из России - ставим ad
3. для источников paid и email не из России - ставим other
4. все остальные варианты берем из traffic_source без изменений

```python
df = log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
log['traffic_type'] = log['traffic_source']
log.loc[((log.traffic_source == 'yandex') | (log.traffic_source == 'google')), 'traffic_type'] = 'organic'
log.loc[(((log.traffic_source == 'paid') | (log.traffic_source == 'email')) & (log.region == 'Russia')), 'traffic_type'] = 'ad'
log.loc[(((log.traffic_source == 'paid') | (log.traffic_source == 'email')) & (log.region != 'Russia' )), 'traffic_type'] = 'other'
```
